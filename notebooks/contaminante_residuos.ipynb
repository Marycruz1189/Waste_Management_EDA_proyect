```python
emisiones = pd.read_excel(csv_fullpath)
```

```python
# creo una nueva columna de Total Nacional 

col_list= list(emisiones)
col_list.remove('ANNO')
col_list


emisiones['Total Nacional'] = emisiones[col_list].sum(axis=1)
```

```python
emisiones
```

```python
emisiones = emisiones.drop(columns=['SECTOR','ACTIVIDAD'], axis=1)
```

```python
to_drop = emisiones.iloc[:, 2:30]
emisiones.drop(to_drop, inplace = True , axis= 1 )
```

```python
emisiones
```

```python
#voy a cambiar el nombre de la columna año 
emisiones.rename(columns={'ANNO': 'Año'}, inplace= True) 
```

```python
emisiones['Total Nacional'] = pd.Series([round(val, 2) for val in emisiones ['Total Nacional']])
emisiones['Total Nacional']
```

```python
pivot_contami = emisiones.pivot(index='Año', columns='DESCRIPCION', values='Total Nacional')
```

```python
pivot_contami
```

```python
#CREO LAS DOS COLUMNAS QUE ME HACEN FALTA QUE SON AÑOS Y COMUNIDADES AUTONOMAS, Y REINICIO EL INDEX
pivot_contami['Año'] = pivot_contami.index
```

```python
pivot_contami['Comunidad Autonoma'] = 'Total Nacional'
```

```python
pivot_contami.reset_index(inplace=True, drop=True)
```
